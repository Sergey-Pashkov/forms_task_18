### Домашнее задание: Разработка многофункциональной формы с использованием Django

#### Задание:

Вы должны создать веб-приложение с формами, которое включает функциональность регистрации, авторизации и отправки сообщений. Проект должен включать следующие компоненты:

1. **Создание модели пользователя (Обязательно)**:
   - Используйте встроенную модель `User` Django или создайте свою модель пользователя с дополнительными полями (например, `phone_number`, `bio`).

2. **Создание форм (Обязательно)**:
   - Создайте следующие формы с использованием `forms.ModelForm`:
     - **Форма регистрации**: включает поля для ввода имени пользователя, электронной почты, пароля и подтверждения пароля.
     - **Форма авторизации**: включает поля для ввода имени пользователя и пароля.
     - **Форма отправки сообщения**: включает поля для ввода имени, электронной почты и текста сообщения.

3. **Валидация форм (Обязательно)**:
   - Реализуйте кастомную валидацию для формы регистрации, чтобы проверить, совпадают ли пароли и чтобы электронная почта не была уже зарегистрирована.
   - Для формы отправки сообщения проверьте, что текст сообщения не пустой и соответствует заданным условиям (например, не превышает 500 символов).

4. **Обработка форм (По желанию)**:
   - Реализуйте представления для обработки всех форм (регистрация, авторизация и отправка сообщения).
   - После успешной регистрации пользователя перенаправляйте его на страницу с сообщением об успешной регистрации.
   - После успешной авторизации перенаправляйте пользователя на главную страницу (например, `/home`).
   - После отправки сообщения выводите сообщение об успешной отправке и возвращайте пользователя на страницу отправки сообщения.

5. **Ajax для отправки сообщений (По желанию)**:
   - Реализуйте отправку формы отправки сообщения с использованием Ajax без перезагрузки страницы.
   - Обработайте ответ от сервера и выведите сообщения об успехе или ошибках валидации на страницу.

6. **Создание шаблонов (По желанию)**:
   - Создайте шаблоны для каждой формы с соответствующим стилем (используйте CSS или Bootstrap).
   - Реализуйте страницу профиля пользователя, где отображаются его данные и отправленные сообщения.

7. **Реализация маршрутов (Обязательно)**:
   - Настройте маршруты (URLs) для всех созданных представлений.

8. **Документация (По желанию, но желательно)**:
   - Напишите документацию к вашему проекту, описывающую, как установить и запустить приложение, а также объясняющую основные функции.
